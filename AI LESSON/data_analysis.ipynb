```python
import pandas as pd 
data = {
    'Months':['Jan','Feb','Mar','Apr','May','Jun'],
    'Sales': [10000,12000,15000,13000,17000,16000],
    'Profit':[2500,2700,3000,2800 ,3500,3200],
    
}
df = pd.DataFrame(data)
```

```python
df
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(8,5)) # 8 is width and 5 is heaight 
plt.plot(df['Months'],df['Sales'], color = 'blue', marker ='o',linestyle= '--', label ='Sales')
plt.title('Sales Trend Over Months')
plt.xlabel('Sales($)')
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
# bar plot SAles vs Profit by Month
plt.figure(figsize=(10,6))
width= 0.3
plt.bar(df['Months'], df['Sales'], width= width, label = 'Sales', color = 'skyblue')
plt.bar(df['Months'],df['Profit'], width= width, label='Porfit',color= 'orange', bottom=df['Sales'])
plt.title('Sales and Profit comparison by month')
plt.xlabel('Months')
plt.ylabel('Amount($)')
plt.legend()
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize = (7,7))
plt.pie(df['Profit'],labels =df['Months'], autopct ='%1.1f%%',startangle=140,colors=plt.cm.Paired.colors)
plt.title('Profit Distribution by Month')
plt.tight_layout()
plt.show()
```

```python
#scatter plot : Sales vs Profit

plt.figure(figsize = (8,5))
plt.scatter(df['Sales'], df['Profit'],color = 'green', s= 100, edgecolor= 'black')
plt.title('Sales vs Profit')
plt.xlabel('Sales($)')
plt.ylabel('Profit($)')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# box plot for profit distribution

plt.figure(figsize=(8,5))
plt.boxplot(df['Profit'], vert=True, patch_artist=True, boxprops=dict(facecolor= 'lightgreen'))
plt.title('profit Distribution')
plt.xlabel('Profit($)')
plt.tight_layout()
plt.show()
```

```python
import gradio as gr
import pandas as pd 
import matplotlib.pyplot as plt
#import numpy as np

data = {
    'Months':['Jan','Feb','Mar','Apr','May','Jun'],
    'Sales': [10000,12000,15000,13000,17000,16000],
    'Profit':[2500,2700,3000,2800 ,3500,3200],
    
}
df = pd.DataFrame(data)

def generate_plot(plot_type):
    plt.figure(figsize=(8,5))
    if plot_type == 'Line Plot':
        plt.plot(df['Months'], df['Sales'], color='blue', marker='o', linestyle='--', label='Sales')
        plt.title('Sales Trend Over Months')
        plt.xlabel('Months')
        plt.ylabel('Sales ($)')
        plt.grid(True)
        plt.legend()
          
    elif plot_type == 'Bar Chart':
        width = 0.3
        plt.bar(df['Months'], df['Sales'], width=width, label='Sales', color='skyblue')
        plt.bar(df['Months'], df['Profit'], width=width, label='Profit', color='orange', bottom=df['Sales'])
        plt.title('Sales and Profit comparison by Month')
        plt.xlabel('Months')
        plt.ylabel('Amount ($)')
        plt.legend()
        
    elif plot_type =='Pie Chart':
        plt.pie(df['Profit'],labels =df['Months'], autopct ='%1.1f%%',startangle=140,colors=plt.cm.Paired.colors)
        plt.title('Profit Distribution by Month')

    elif plot_type == 'Scatter Plot':
        plt.scatter(df['Sales'], df['Profit'],color = 'green', s= 100, edgecolor= 'black')
        plt.title('Sales vs Profit')
        plt.xlabel('Sales($)')
        plt.ylabel('Profit($)')
        plt.grid(True)

    elif plot_type == 'Box Plot':
        plt.boxplot(df['Profit'], vert=True, patch_artist=True, boxprops=dict(facecolor= 'lightgreen'))
        plt.title('profit Distribution')
        plt.xlabel('Profit($)')


    plt.tight_layout()
    return plt


demo = gr.Interface(
    fn = generate_plot,
    inputs = gr.Dropdown(choices = ['Line Plot', 'Bar Chart', 'Pie Chart','Scatter Plot','Box Plot'] , label = 'Select Plot Type'),
    outputs = gr.Plot(label = 'Gernerated Plot'),
    title = 'Dynamic Date Visualizatio with gradio',
    description = 'Select a plot type to visualize the sales and profit data'
)
demo.launch()
        
        
```

```python
import seaborn as sns 
import matplotlib.pyplot as plt 

# load sample dataset
tips = sns.load_dataset("tips")

#Set a visual theme
sns.set_theme(style= "darkgrid")
```

```python
plt.figure(figsize=(8,6))
```

```python
sns.scatterplot(data =tips , x="total_bill", y= "tip", hue = "time",size = "size", palette = "deep")
plt.title("Scatter Plot: Total Bill vs Tip by Time and Size")
plt.show()
```

```python
sns.barplot(data=tips , x= "day", y="total_bill", hue ="sex", palette = "muted")
plt.title("Bar Plot : Average Total Bill by Day and Sex")
plt.show()
```

```python
sns.lineplot(data = tips , x= "size", y= "total_bill", hue = "sex", ci=None , marker= "o")
plt.title("Line Plot: Total bill by Party Size and Sex")
plt.show()
```

```python
sns.boxplot(data = tips , x= "day", y = "tip", hue="smoker", palette = "Set2")
plt.title("Box Plot: Tips by Day and Smoke Status")
plt.show()
```

```python
# Violine Plot: Total Bill By Day ,split by time

sns.violinplot(data = tips , x= "day", y ="total_bill" , hue="time",split= True ,palette ="pastel")
plt.title("Violin Plot: Totla Bill by Day and Time")
plt.show()
```

```python
#Count plot: Order by Day ,split by smoker

sns.countplot(data = tips, x = "day", hue = "smoker", palette = "dark")
plt.title("Count plot: Orders by Day and Smoker Status")
plt.show()
```

```python
sns.regplot(data= tips, x= "total_bill", y="tip", scatter_kws={"s": 50}, line_kws = {"color": "red"})
plt.title("Regressuib Plot: Totla Bill vs Tip")
plt.show()
```

```python
# Histogram(Distribution Plot): totla Bill with KDE

sns.histplot(data = tips , x= "total_bill", kde = True, bins=20,color="blue")
plt.title("Histogram: Distribution of Total Bill")
plt.show()
```

```python
#9 Pair Plot: Relationships between numerical variables

sns.pairplot(tips, hue= "sex", vars =["total_bill", "tip", "size"], palette = "husl")
plt.suptitle("Pair Plot: Numerical Variables by Sex", y= 1.02)
plt.show()
```

```python
# catplot(Point plot ): Tip by Day and Sex
sns.catplot(data = tips , x= "day", y = "tip" , hue = "sex", kind = "point",palette = "bright")
plt.title("Catplot(Point): Tips by Day and Sex")
plt.show()
```

```python
# Joint Plot : total Bill vs Tip with marginal distributions
sns.jointplot(data =tips , x="total_bill", y = "tip", kind = "scatter", hue="smoker", palette = "coolwarm")
plt.suptitle("Joint Plot: Total Bill vs Tip by Smoker", y=1.02)
plt.show()
```

```python
# FacetGrid: Total Bill by Day, faceted by time and smoker
g = sns.FacetGrid(tips, col= "time", row= "smoker", margin_titles=True)
g.map(sns.histplot, "total_bill", bins=15)
g.fig.suptitle("FaceGrid: Total Bill by Time and Smoker", y=1.02)
plt.show()
```

```python
# Strip Plot: Tips by day , colored by sex

plt.figure(figsize = (8,6))
sns.stripplot(data= tips, x="day", y="tip", hue="sex", palette= "Set1", jitter= True)
plt.title("Strip Plot: Tips by Day and sex")
plt.show()
```

```python
# KDE  Plot: Total Bill density by sex
plt.figure(figsize=(8,6))
sns.kdeplot(data = tips , x = "total_bill", hue = 'sex',fill = True , palette ="tab10")
plt.title("KDE  Plot: Total Bill Density by sex")
plt.show()
```

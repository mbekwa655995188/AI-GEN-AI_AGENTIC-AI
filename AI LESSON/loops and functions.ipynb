```python
#positional argument
def person(name, age):
    print(name)
    print(age)
person('lemuel',33)
```

```python
def person(name,age):
    print(name)
    print(age)

person(40,'lemuel')
```

```python
# keyword argument is to overcome position argument
def person(name, age , newage):
    print(name)
    print(age)
    print(newage)
person(age=34, name = 'Lemuel', newage = 55)
```

```python
# default argument
def person(name , age = 18):
    print (name)
    print(age)
person('Elvina')
```

# Variable Length argument

```python
def sum(a,b):
    c = a+b 
    print(c)
sum(3,5)
```

```python
def sum(a,*b):
    c = a+ b 
    print(c)
sum(3,4,5,6)
```

```python
def sum(a,*b):
    c = a
    for i in b:
        c = c + i 
    print(c)
sum(3,4,5,6)
     
```

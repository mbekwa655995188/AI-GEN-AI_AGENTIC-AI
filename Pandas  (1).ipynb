# visualisation  -- matplotlib and seaborn.

```python
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline # graph size
plt.rcParams['figure.figsize'] =  6 , 2     
warnings.filterwarnings('ignore')  # even as are update please ignore all the warning messages
```

```python
df['InternetUsers']
```

```python
 vis1 =sns.distplot(df['InternetUsers'] ,bins=10)
plt.show()
```

```python
vis2 = sns.distplot(df["InternetUsers"] , bins=10)
plt.show()
```

```python
vis3 = sns.boxplot(data = df , x='IncomeGroup', y ='BirthRate')
plt.show()
```

```python
vis4 = sns.lmplot(data = df , x='InternetUsers', y ='BirthRate')
plt.show() 
```

```python
vis5 = sns.lmplot(data =df, x = 'InternetUsers' , y='BirthRate' , fit_reg = False)
plt.show()
```

```python
vis6 = sns.lmplot(data =df , x= 'InternetUsers' , y='BirthRate' , fit_reg = True , hue = 'IncomeGroup')
plt.show()
```
